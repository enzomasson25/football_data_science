```python

def calculMoyAge(tabAgeCplt,tabValCplt):
    age = []
    moy = []
    for i in range(0,len(tabAgeCplt)):
        if tabAgeCplt[i] not in age:
            age.append(tabAgeCplt[i])
            moy.append([tabValCplt[i],1])
        else:
            rang = age.index(tabAgeCplt[i])
            moy[rang] = [moy[rang][0]+tabValCplt[i],moy[rang][1]+1]
    moyFinale = []
    for elem in moy:
        moyFinale.append(int(elem[0]/elem[1]))
    
    return age,moyFinale
        
valAge,valpot = calculMoyAge(age,pot)


plt.scatter(valAge,valpot)
plt.xlabel('age')
plt.ylabel('potential')
plt.title('Moyenne du potentiel en fonction de l\'age')
plt.show()
```

```python
def calculMoyAgePoste(tabAgeCplt,tabValCplt,tabPosCplt):
    ageAndPosi = []
    moy = []
    for i in range(0,len(tabAgeCplt)):
        if [tabAgeCplt[i],tabPosCplt[i]] not in ageAndPosi:
            ageAndPosi.append([tabAgeCplt[i],tabPosCplt[i]])
            moy.append([tabValCplt[i],1])
        else:
            rang = ageAndPosi.index([tabAgeCplt[i],tabPosCplt[i]])
            moy[rang] = [moy[rang][0]+tabValCplt[i],moy[rang][1]+1]
    moyFinale = []
    for elem in moy:
        moyFinale.append(int(elem[0]/elem[1]))
    
    for i in range(0,len(ageAndPosi)):
        ageAndPosi[i].append(moyFinale[i])
    
    return ageAndPosi  


def parPoste(tab,poste):
    tabPoste = []
    for i in range(len(tab)):
        if tab[i][1] == poste:
            tabPoste.append(tab[i])
    return tabPoste


import numpy as np

x = calculMoyAgePoste(age,phy,position)
goal = parPoste(x,'GK')
dd = parPoste(x,'RB')
dld = parPoste(x,'RWB')
dg = parPoste(x,'RG')
dlg = parPoste(x,'LWB')
dc = parPoste(x,'CD')

truc = [i[0] for i in x]
plt.scatter([i[0] for i in goal], [i[2] for i in goal], s = 130, c = 'yellow', marker = '.', edgecolors = 'green', label="goal")
plt.scatter([i[0] for i in dd], [i[2] for i in dd], s = 130, c = 'red', marker = '.', edgecolors = 'blue',label="dd")
plt.scatter([i[0] for i in dld], [i[2] for i in dld], s = 130, c = 'black', marker = '.', edgecolors = 'red',label="dld")
plt.scatter([i[0] for i in dg], [i[2] for i in dg], s = 130, c = 'blue', marker = '.', edgecolors = 'green',label="dg")
plt.scatter([i[0] for i in dlg], [i[2] for i in dlg], s = 130, c = 'purple', marker = '.', edgecolors = 'green',label="dlg")
plt.scatter([i[0] for i in dc], [i[2] for i in dc], s = 130, c = 'orange', marker = '.', edgecolors = 'green',label="dc")



plt.xlabel('age')
plt.ylabel('phy')
plt.legend()
plt.title('Moyenne note physique des défenseurs en fonction de l\'age ')
plt.show()
```
